### Step 2: Scrape the Table

Create a Beautiful Soup object and use it to scrape the data in the HTML table.

Note that this can also be achieved by using the Pandas `read_html` function. However, use Beautiful Soup here to continue sharpening your web scraping skills.

```python
# Create a Beautiful Soup Object
html = browser.html
w_soup = soup(html, 'html.parser')
```

```python
weather_headers = w_soup.find_all('th')
weather_data = w_soup.find_all('tr', class_='data-row')
```

```python
# confirming data was retrieved and can be manipulated
print("Data Rows: ", len(weather_data))
for row in weather_data:
    print(row.text)
```

### Step 3: Store the Data

Assemble the scraped data into a Pandas DataFrame. The columns should have the same headings as the table on the website. Here’s an explanation of the column headings:

* `id`: the identification number of a single transmission from the Curiosity rover
* `terrestrial_date`: the date on Earth
* `sol`: the number of elapsed sols (Martian days) since Curiosity landed on Mars
* `ls`: the solar longitude
* `month`: the Martian month
* `min_temp`: the minimum temperature, in Celsius, of a single Martian day (sol)
* `pressure`: The atmospheric pressure at Curiosity's location

```python
# Create an empty list
weather = []
# Loop through the scraped data to create a list of rows
for row in weather_data:
    weather.append(row.text)
weather
```

```python
# Create a Pandas DataFrame by using the list of rows and a list of the column names
weather_list = pd.read_html(html, header=0, flavor='bs4', index_col='id')
# the line above creates a 'list' object, and satisfies the challenge requirements as worded.
# I could use df[["id", "terrestrial_date", "sol", "ls", "month", "min_temp", "pressure"]] = df["0"].str.split('\n', n=1, expand=True)
# but that is unwieldy at best and an active hindrance at worst. I took more time on this comment than getting the DataFrame.
weather_df = weather_list[0]
# the line above converts the df list item to a true DataFrame
```

### Step 5: Analyze the Data

Analyze your dataset by using Pandas functions to answer the following questions:

1. How many months exist on Mars?
2. How many Martian (and not Earth) days worth of data exist in the scraped dataset?
3. What are the coldest and the warmest months on Mars (at the location of Curiosity)? To answer this question:
    * Find the average the minimum daily temperature for all of the months.
    * Plot the results as a bar chart.
4. Which months have the lowest and the highest atmospheric pressure on Mars? To answer this question:
    * Find the average the daily atmospheric pressure of all the months.
    * Plot the results as a bar chart.
5. About how many terrestrial (Earth) days exist in a Martian year? To answer this question:
    * Consider how many days elapse on Earth in the time that Mars circles the Sun once.
    * Visually estimate the result by plotting the daily minimum temperature.


```python
# 1. How many months are there on Mars?
num_months = weather_df['month'].unique()
print(f'There are {len(num_months)} unique months, based on the data provided (though they are longer).\nMartian months change length regularly.')
```

```python
# 2. How many Martian days' worth of data are there?
num_days = weather_df['sol'].count()
print(f'There are {num_days} Earth days of data available,\nor {(num_days * 1.02749125):.2f} Martian Days,\nwhich is {(num_days / 687):.2f} Martian Years.')
```

```python
# 3. What is the average low temperature by month?
avg_df_month = weather_df.groupby('month').mean()
avg_low_temp_min = avg_df_month['min_temp'].min()
avg_low_temp_max = avg_df_month['min_temp'].max()
# print(avg_low_temp_min, avg_low_temp_max)
avg_df_month['min_temp']
```

```python
# Plot the average temperature by month
x_axis = num_months
y_axis = avg_df_month['min_temp']
plt.bar(x_axis, y_axis, color='red', label='Martian Low Temps')
plt.xlabel("Martian Month")
plt.xticks(np.arange(0, 13, step=1.0))
# plt.yticks(np.arange(1, -70, step=5.0)) # would not generate yticks
plt.ylabel("Daily Average Low Temp")
plt.title("Martian Monthly Lows: Average")
plt.figure(figsize=(10, 50))
```

```python
# Identify the coldest and hottest months in Curiosity's location
# get temps from the df
min_temp = avg_df_month['min_temp'].min()
max_temp = avg_df_month['min_temp'].max()
# find the rows with associated data
min_temp_month = avg_df_month.loc[avg_df_month['min_temp'] == min_temp]
max_temp_month = avg_df_month.loc[avg_df_month['min_temp'] == max_temp]
# show results
print(min_temp_month, max_temp_month)
```

```python
# 4. Average pressure by Martian month
avg_pressure = weather_df.groupby(['month']).mean()
avg_month_pressure = avg_pressure['pressure']
avg_month_pressure
```

```python
# Plot the average pressure by month
x_axis = num_months
y_axis = avg_month_pressure
plt.bar(x_axis, y_axis, color='red', label='Martian Pressure')
plt.xlabel("Month")
plt.xticks(np.arange(0, 13, step=1.0))
plt.ylabel("Daily Average Pressure (atm)")
plt.title("Martian Monthly Pressure: Average")
plt.figure(figsize=(10, 50))
```

```python
# 5. How many terrestrial (earth) days are there in a Martian year?
print(f'There are {num_days} Earth days of data available,\nwhich is {(num_days / 687):.2f} Martian Years.\nWhen divided, we get 686.39 days, close to expected.')
```

On average, the third month has the coldest minimum temperature on Mars, and the eighth month is the warmest. But it is always very cold there in human terms! Atmospheric pressure is, on average, lowest in the sixth month and highest in the ninth. The distance from peak to peak is roughly 1425-750, or 675 days. A year on Mars appears to be about 675 days from the plot. Internet search confirms that a Mars year is equivalent to 687 earth days.
